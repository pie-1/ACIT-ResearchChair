```python
#Filtering = Refers to the process of selecting elements 
#            from an array that match a given conditions
import numpy as np
```

```python
ages = np.array([[21, 17, 19, 20, 16, 30, 18, 65],
                 [39, 22, 15, 67, 18, 19, 20, 21]])
```

```python
teenagers = ages[ages < 18]
print(teenagers)
```

```python
adults = ages[(ages >= 18) & (ages < 65)]            #in python logical and is 'and'
print(adults)                               #but since numpy is written in C, we use & as the logical operator
```

```python
seniors = ages[ages >= 65]
print(seniors)
```

```python
evens = ages[ages % 2 ==0]
print(evens)
```

```python
odds = ages[ages % 2 != 0]
print(odds)
```

```python
#where function is used to preserve the structure/shape of the array
adults = np.where(ages >= 18, ages, 0)    #1st argument = condition, 2nd arg = array, 3rd arg is a fill value!!!
print(adults)
```
