```python
import numpy as np
```

```python
# we need random numbers for simulation? This is the method to generate them.
rng = np.random.default_rng()       #rng = random number generator
                                    #Note: We must setup a random number generator
```

random numbers when rolling a dice

```python
print(rng.integers(1, 7))   #(1, 7) is the range from 1 to less then 7 
#We get random values as if we rolled a dice
```

```python
print(rng.integers(low = 1, high = 101))  # for better readibility
```

```python
print(rng.integers(low = 1, high = 101, size = 6))
#We get 6 probable numbers between 1 and 100
```

```python
# If I need to store them in a 2d array, I write
print(rng.integers(low = 1, high = 101, size = (3, 3)))
```

```python
rng = np.random.default_rng(seed = 1)                    #seed is used if we have to reproduce a number
print(rng.integers(low = 1, high = 101, size = (3, 3))) #If we run this again, we will get same result
```

For floating point numbers

```python
print(np.random.uniform(low = -1, high = 1, size = (3, 2)))
```

```python
np.random.seed(seed = 1)
print(np.random.uniform(low = -1, high = 1, size = (3, 2)))
```

Shuffle in array

```python
rng = np.random.default_rng()
array = np.array([1, 2, 3, 4, 5])
rng.shuffle(array)
print(array)
```

```python
# if we need a choice
fruits = np.array(['apple', 'orange', 'banana', 'coconut'])
fruits = rng.choice(fruits)
print(fruits)
#if I want to eat 3 fruits
fruits = np.array(['apple', 'orange', 'banana', 'coconut'])
fruits = rng.choice(fruits, size =(3, 3))
print(fruits)
```

For selecting emojis windows + ;

```python
emotion = np.array(['😊', '😂', '😘', '😒', '😢'])   
emotion = rng.choice(emotion, size =(3, 3))
print(emotion)
```
