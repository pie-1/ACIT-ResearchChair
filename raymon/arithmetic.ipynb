```python
import numpy as np
```

```python
#Scalar arithmetic
array = np.array([1, 2, 3])
```

```python
print(array)
print(array.shape) #Output: (3,) means The array has 3 elements. It is 1-dimensional.
print(array.ndim)
print(array.dtype)
```

```python
print(array + 1)
print(array - 1)
```

```python
print(array * 2)
print(array / 2)
```

```python
# For square multiplication
print(array ** 2)
```

For Vector arithmetics

```python
# We use vectorized math function
print(np.sqrt(array))
```

```python
array2 = np.sqrt(array)
print(array2)
print("To round each float elements to nearest integer we use the round function")
print(np.round(array2))
print("To round up to nearest bigger integer we use the ceil function")
print(np.ceil(array2))
print("To round down to nearest smaller integer we use the floor function")
print(np.floor(array2))
```

```python
print(np.pi)     #pi value
```

```python
radii = np.array([1, 2, 3])
```

```python
print(np.pi * radii ** 2)               #area of circle
```

```python
# Element wise arithmetic
array3 = np.array([1, 2, 3]) 
array4 = np.array([4, 5 ,6])
print(array3 + array4)
print(array3 / array4)
print(array3 ** array4)
```

```python
# Comparision operators
scores = np.array([91, 55, 100, 73, 82, 64])
print(scores == 100)   #returns boolean value if each element meets criteria
print(scores >= 60)
print(scores < 60)
```

```python
scores[scores < 60] = 0
print(scores)
```
