```python
import pandas as pd
```

```python
print(pd.__version__)
```

```python
#Series = A pandas 1D labeled array that can hold any data type
#         We can think of it like a single column in a spreadsheet (1-D)
intdata = [100, 102, 104, 200, 202]
floatdata = [100.1, 102.2, 104.3]
stringdata = ["raymon", "anjali", "shubham"]
booleandata = [True, False, True]
```

```python
series = pd.Series(intdata)
print(series)
```

```python
floatseries = pd.Series(floatdata)        
print(floatseries)
```

```python
classmates = pd.Series(stringdata)        
print(classmates)
```

```python
booleanseries = pd.Series(booleandata)        
print(booleanseries)
```

```python
#We can also assign a unique index for each array element
classmates = pd.Series(stringdata, index =["boy","girl","boy"])      #The index can also be a tuple, dictionary, another numpy array another series
print(classmates)
```

```python
#if we need to access a particular element inside a series we use loc method;
#To print:
print(classmates.loc["boy"])   
#To print only raymon:
print(classmates.iloc[0])
```

```python
print(classmates.loc["girl"])
```

```python
print(series.loc[0])
```

```python
#To change a particular element:
classmates.loc["boy"] = "Bijaya"
print(classmates)               #Which isn't desirable
#                                since I want 1st boy to be raymon

```

```python
#To solve this we use the iloc method which ignores the text labels entirely 
# and looks at the 0-based index count (0, 1, 2...).
classmates.iloc[0] = "raymon"
print(classmates)
```

```python
#Filtering data based on conditions
series = pd.Series(intdata, index = ('a','b','c','d','e'))
print(series[series >= 200])
```

```python
# a more complex example:
#lets create a series to track my diet for 3 days
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}   #This is a dictionary, which assigns a unique index to a element automatically
series = pd.Series(calories)
print(series)
print(series.loc["Day 3"])
```

```python
print(series.loc["Day 3"])
#Lets say i cheated on day 3 and had a extra burger
series.loc["Day 3"] += 1500
print(series.loc["Day 3"])
```

```python
#lets find the day where i have not followed my diet plan of not eating more than 2000 calories
print(series[series > 2000])
```
