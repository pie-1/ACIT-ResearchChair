 DataFrame in pandas, they are tabulated data structure with rows and columns. Similar to spreadsheet like excel.

```python
import pandas as pd
```

```python

data = {"Name": ["spongebob", "patrick", "squidward"],
        "Age": [30, 35, 50]
}  #This is a dictionary
```

```python
df = pd.DataFrame(data)     #index is 0, 1, 2 by default
print(df)
```

```python

df = pd.DataFrame(data, index = ["Employee 1", 'Employee 2', 'Employee 3']) #df = dataframe
#                                  in DataFrame() is a constructor not a functions
print(df)
print(df.loc["Employee 1"])  # to access a particular individual.
```

```python
#Adding a new column to a dataframe.
df["Job"] = ["Cook", "N/A", "Cashier"]
print(df)
```

```python
df["Salary"] =["Absolutely nothing", "He owns mr krabs money", 10000]
print(df)
```

```python
# Both list and tuples are heterogenous data structures
list = [10, 'a', 2.333]         # A daily remainder.
print(list)
tuple = (10.2, 10, 'a')
print(tuple)
```

Adding a new rows to a dataframe:

```python
new_row = pd.DataFrame([{'Name':'Sandy', 
                         "Age": 28,
                         "Job": "Engineeer"
                         }, {'Name':'Ryomen Sukuna', 
                         "Age": 1000,
                         "Job": "Vegetable fillet and dissect"
                         }], index=["Employee 4", "Employee 5"])

#           Remember to create a dataframe we need a python dictionary.
#           The reason why we are also passing index will come later below.
```

We create a new row in dataframe by creating a new smaller dataframe and concatinate it to the original.
NOTE: The use of '.append' to concatenate a DataFrame to another DataFrame is deprecated. 
Use '.concat' instead.

```python
df = pd.concat([df, new_row]) #Syntax: We pass in the list of all dataframes we want to concatinate.
#                                      as the argument for .concat
print(df)
```
